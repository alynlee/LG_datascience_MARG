# 옵션 1 : 구글 드라이브로 음원 직접 업로드 가능한지 확인

음원 공유 링크

https://drive.google.com/open?id=1EyFGkYdWqXE1IgDQKQ6uDsk_7tNwKuRY

```
import librosa
import IPython.display as ipd


audio_path = '/content/drive/My Drive/musicOrig.wav'

audio, sr = librosa.load(audio_path, sr=None)
print('sample rate of this audio file :', sr, 'Hz')
ipd.Audio(audio, rate=sr, autoplay=False) # play audio
```
